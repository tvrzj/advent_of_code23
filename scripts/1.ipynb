```python
import os
import re
```

```python
data = []
a = open('../data/1input').readlines()
for line in a:
    data.append(line.rstrip('\n'))
data
```

```python
numbers = []
for line in data:
    numbers.append(re.findall('\d', line))

numbers
```

```python
result = []
for line in numbers:
    joined_inner_list = int(line[0] + line[-1])
    result.append(joined_inner_list)

sum(result)
```

# Part2:

```python
digit_map = {
        'one': 'o1e',
        'two': 't2o',
        'three': 't3e',
        'four': 'f4r',
        'five': 'f5e',
        'six': 's6x',
        'seven': 's7n',
        'eight': 'e8t',
        'nine': 'n9e',
        }
```

```python
data2 = []
for item in data:
    for word, digit in digit_map.items():
        item = item.replace(word, digit)
    data2.append(item)

data2[0:5]
```

```python
numbers2 = []
for line in data2:
    numbers2.append(re.findall('\d', line))

numbers2[0:5]
```

```python
result2 = []
for line in numbers2:
    joined_inner_list = int(line[0] + line[-1])
    result2.append(joined_inner_list)

sum(result2)
```

```python
import pandas as pd
sanity_df = pd.DataFrame(data=[data, numbers, result, data2, numbers2, result2], index=('data', 'numbers', 'result', 'data2', 'numbers2', 'result2')).transpose()
sanity_df
```
